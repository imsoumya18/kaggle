```python
import pandas as pd
```

```python
data = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
test_ids = test["Id"]
```

```python
# dropping columns
data.drop(["Id"], axis=1)
test.drop(["Id"], axis=1)
```

```python
# check if there is NaN
data.columns[data.isna().any()].tolist()
```

```python
# fill NaN
cols_na = ["LotFrontage", "GarageYrBlt"]

for col in cols_na:
    data[col].fillna(data[col].median(), inplace=True)
    test[col].fillna(test[col].median(), inplace=True)
```

```python
# label encoding
import sklearn.preprocessing as pre

le = pre.LabelEncoder()

cols_label = ["MSSubClass", "MSZoning", "Street", "Alley", "LotShape", "LandContour", "Utilities", "LotConfig", "LandSlope", "Neighborhood", "Condition1", "Condition2",
              "BldgType", "HouseStyle", "RoofStyle", "RoofMatl", "Exterior1st", "Exterior2nd", "MasVnrType", "ExterQual", "ExterCond", "Foundation", "BsmtQual", "BsmtCond",
              "BsmtExposure", "BsmtFinType1", "BsmtFinSF1", "BsmtFinType2", "BsmtFinSF2", "Heating", "HeatingQC", "CentralAir", "Electrical", "KitchenQual", "Functional",
              "FireplaceQu", "GarageType", "GarageFinish", "GarageQual", "GarageCond", "PavedDrive", "PoolQC", "Fence", "MiscFeature", "SaleType", "SaleCondition"]

for col in cols_label:
    data[col] = le.fit_transform(data[col])
    test[col] = le.transform(test[col])
    print(le.classes_)
```
